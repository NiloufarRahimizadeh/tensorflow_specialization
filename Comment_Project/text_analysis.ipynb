```python
import nltk
import pandas as pd
```

```python
data_path = '/Users/niloufar/Desktop/DeepLearning/tf_specialization/comment/'
data1 = 'spam_or_not1.xlsx'
data2 = 'spam_or_not2.xlsx'
data3 = 'spam_or_not3.xlsx'
df1 = pd.read_excel(data_path + data1)
df2 = pd.read_excel(data_path + data2)
df3 = pd.read_excel(data_path + data3)
df = pd.concat([df1, df2, df3], ignore_index=True)
df = df.drop(['ID', df.columns[3]], axis=1)
# df.iloc[995:1005, :]
df.head()
```

```python
comments = []
for i in df['Text (comment)']:
    comments.append(i)
```

```python
len(comments)
```

```python
moby_tokens = nltk.word_tokenize(str(comments))
```

```python
def example_one():
    return len(moby_tokens)
example_one()
```

```python
def example_two():
    return len(set(moby_tokens))

example_two()
```

```python
def answer_three():
    fdist = nltk.FreqDist(moby_tokens)
    return fdist.most_common()
answer_three()
```
